## Pixel masking

```python
theta, ra = healpy.pix2ang(128, np.arange(healpy.nside2npix(128)))
dec = np.pi/2 - theta
```

```python
# simple example mask
mask = (
    (np.deg2rad(120) <= ra) & (ra <= np.deg2rad(240))
    & (np.deg2rad(-30) <= dec) & (dec <= np.deg2rad(30))
)
mask.mean()
```

```python
sstr5_mask = cy.get_sky_scan_trial_runner(ana=ana, src_kw=dict(extension=np.deg2rad(5)), pixmask=mask)
```

```python
%%time

scan5_mask = sstr5_mask.get_one_scan(seed=1, logging=True, mp_cpus=20)
```

```python
healpy.mollview(scan5_mask[1], coord='C', rot=180, max=19.2)
```
